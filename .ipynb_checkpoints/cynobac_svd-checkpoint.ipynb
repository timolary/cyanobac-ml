```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
%matplotlib inline
import sklearn
```

### Current samples times above the threshold

```python
ds1 = pd.read_csv('data/AllSites.csv')
threshold = np.where(ds1['NP_Cya_bio'] >= 4e8)
target = np.zeros(len(ds1['NP_Cya_bio']))
target[threshold] = 1
ds1['target'] = pd.Series(target)
ds1 = ds1.dropna(axis=0, how='any')
```

```python
t = np.where(ds1['target'] == 1.0)
ds1.iloc[t]
```

### This almost works... it rounds some of the floats too much, will continue to address.

```python
ds2 = ds1.drop(['Station', 'Stratum','Date','StationID','Time'], axis=1)
# ds2
ds2['Depth'] = ds1['Depth'].astype(str).str.extract('(\d+)').astype(float)
ds2['TP'] = ds1['TP'].astype(str).str.extract('(\d+)').astype(float)
ds2['Cl'] = ds1['Cl'].astype(str).str.extract('(\d+)').astype(float)
ds2['DP'] = ds1['DP'].astype(str).str.extract('(\d+)').astype(float)
ds2['TN'] = ds1['TN'].astype(str).str.extract('(\d+)').astype(float)
ds2['TempC'] = ds1['TempC'].astype(str).str.extract('(\d+)').astype(float)
ds2['Chla'] = ds1['Chla'].astype(str).str.extract('(\d+)').astype(float)
ds2['Chla'] = ds1['Chla'].astype(str).str.extract('(\d+)').astype(float)
ds2['Secchi'] = ds1['Secchi'].astype(str).str.extract('(\d+)').astype(float)
y = np.array(ds2['target'])
X = np.array(ds2.drop(['target'], axis=1))
# ds1.iloc[t]
ds2
```

### Simple SVM Testing

```python
from sklearn import svm
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import cross_val_score, cross_val_predict

X_normalized = preprocessing.normalize(X) # normalize X for processing
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
print('C=5\n')
for kernel in kernels:
    svm_classifer = svm.SVC(C=5, kernel=kernel)
    y_predicted = cross_val_predict(svm_classifer, X, y, cv=10)
    print(f'KERNEL: {kernel}')
    print(metrics.classification_report(y, y_predicted))
    print(metrics.confusion_matrix(y, y_predicted))
    print('\n')
print('C=50\n')
for kernel in kernels:
    svm_classifer = svm.SVC(C=50, kernel=kernel)
    y_predicted = cross_val_predict(svm_classifer, X, y, cv=5)
    print(f'KERNEL: {kernel}')
    print(metrics.classification_report(y, y_predicted))
    print(metrics.confusion_matrix(y, y_predicted))
    print('\n')

```
